## Problem 16

Determine which sessions occurred on the same day each user registered.

```python
sessions_date_user_registered = pd.merge(left=users, left_on=['UserID', 'Registered'],\
                                         right= sessions, right_on=['UserID', 'SessionDate'], how='inner')
```

```python
sessions_date_user_registered
```

## Problem 17

Build a dataset with every possible (UserID, ProductID) pair (cross join).

```python
users1 = pd.DataFrame({'key': (users.shape[0]), 'UserID': users.UserID})
products1 = pd.DataFrame({'key': (products.shape[0]), 'ProductID': products.ProductID})
userid_prodid_pair = pd.merge(users1, products1, on='key')[['UserID', 'ProductID']]
```

```python
userid_prodid_pair
```

## Problem 18

Determine how much quantity of each product was purchased by each user

```python
qty_product_user = pd.merge( userid_prodid_pair, transactions, how='left', on=['UserID','ProductID']) \
                           [['UserID', 'ProductID', 'Quantity']].fillna(0)
```

```python
qty_product_user
```

## Problem 20

Join each user to his/her first occuring transaction in the transactions table.

```python
user_by_transaction = pd.merge(users, transactions.groupby('UserID').first(), how='left', on='UserID').drop('key', axis=1)
```

```python
user_by_transaction
```

## Problem 21

Test to see if we can drop columns.

```python
my_columns = list(user_by_transaction.columns)
```

```python
list(user_by_transaction.dropna(thresh=int(user_by_transaction.shape[0]*.9), axis=1).columns)
```

```python
missing_info = list(user_by_transaction.columns[user_by_transaction.isnull().any()])
```

```python
missing_info
```

```python
for col in missing_info:
    num_missing = user_by_transaction[user_by_transaction[col].isnull() == True].shape[0]
    print('number missing for column {}: {}'.format(col, num_missing))
```

```python
for col in missing_info:
    percent_missing = user_by_transaction[user_by_transaction[col].isnull() == True].shape[0]/user_by_transaction.shape[0]
    print('percent missing for column {}: {} percent'.format(col,percent_missing))
```
